```python
data = pd.read_csv("../input/netflix-original-films-imdb-scores/NetflixOriginals.csv",encoding='latin-1')
```

# *****2019 OCAK İLE 2020 HAZİRAN TARİHLERİ ARASINDA 'Documentary' TÜRÜNDE ÇEKİLMİŞ FİLMLERİN IMDB DEĞERLERİNİ BULUP GÖRSELLEŞTİRİNİZ*****

```python
premiere=data.iloc[:,2:3]
print(premiere)
genre=data.iloc[:,1:2]
print(genre)
imdb=data.iloc[:,4:5]
print(imdb)

data2=pd.concat([premiere,genre,imdb],axis=1)
print(data2)
```

```python
data["Date"] = pd.to_datetime(data.Premiere)

data_2 = data.loc[(data["Genre"] == "Documentary") & (data["Date"] > "2019-01") & (data["Date"] < "2020-06-01")]
Fig= px.bar(data_2, x=data_2.Title, y = data_2["IMDB Score"])
Fig.show()
```

# IMDB puanı ile 'Runtime' arasında nasıl bir korelasyon vardır inceleyip görselleştiriniz.

```python
data_i=data["IMDB Score"]
data_r=data["Runtime"]
data_7=pd.concat([data_i,data_r],axis=1)
data_i.corr(data_r)
```

```python
corr=np.corrcoef(data_i,data_r)
print(corr)
```

```python

ax = sns.heatmap(corr, linewidth=0.5)
plt.show()
```
